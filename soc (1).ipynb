```python
from datetime import datetime
import pytz
import os

absent_peeps = []

def start():
    print("Enter your name when promtped.")
    print("The time you left will be recorded.")
    print("The time you returned will be recorded.")
    print("When you enter your name the time you left will be marked.")
    print("When you return, press the 'Enter' key to mark the time you returned.")
    print("Enter 'c' to continue.")
    
    answer = input(">")
    
    if answer == 'c':
        os.system("clear")
        run1()
    else:
        os.system("clear")
        run1()

def run1():
    print("")
    print("Returning or leaving, 'l' or 'r'")
    print("")
    answer = input(">")
    
    if answer == "r":
        run2()
    if answer == 'l':
        run3()
    else:
        print("NOT AVAILABLE")
        run1()
    
def run2():
    print("Who is returning?")
    print("Enter 'e' to exit")
    answer = input(">")
    tz_ro = pytz.timezone("America/Chicago")
    current_time = datetime.now(tz_ro)
    
    if answer == "":
        print("")
        print("NOT AVAILABLE")
        print("")
        run2()
    if answer == "e":
        run1()
    if answer in absent_peeps:
        print("Time returned:",current_time.strftime("%T"))
        run1()
    else:
        print("")
        print("NOT AVAILABLE")
        print("")
        run2()

def run3():
    print("Who is leaving?")
    print("Enter 'e' to exit")
    print("")
    answer = input(">")
    tz_ro = pytz.timezone("America/Chicago")
    current_time = datetime.now(tz_ro)
    
    if answer == "":
        print("Enter you're name")
        run3()
    if answer == "e":
        run1()
    else:
        absent_peeps.append(answer)
        print("Time left:",current_time.strftime("%T"))
        run1()
    
    
start()
```
